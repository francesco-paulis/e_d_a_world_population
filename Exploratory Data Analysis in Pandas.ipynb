### sorting on values

```python
# obtaining the less populated countries in 2022

df.sort_values(by = "2022 Population").head()
```

```python
# most pupulated countries in 2022

df.sort_values(by = "2022 Population", ascending = False).head()
```

```python
# the 10 most populated countries in 2022

df.sort_values(by = "2022 Population", ascending = False).head(10)
```

### correlation between columns

```python
df.corr(numeric_only = True) # to obtain a table in which is shown every correlation between all the numeric columns
```

```python
# obtaining a visualization to better get a glimpse of the correlation between columns

sns.heatmap(df.corr(numeric_only = True), annot = True)

plt.rcParams['figure.figsize'] = (15,15)
plt.show()
```

### grouping data to find useful insights

```python
df.groupby('Continent').mean(numeric_only = True)
```

```python
# an in-depth look at a certain category of values

df[df['Continent'].str.contains('Oceania')]
```

```python
df.groupby('Continent').mean(numeric_only = True).sort_values(by = '2022 Population', ascending = False)
```
